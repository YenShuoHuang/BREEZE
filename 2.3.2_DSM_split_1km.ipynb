## 1. Prepare the city mask vector file
### (1) Find an ETRS89 vector file

find a vector file with ETRS89 (EPSG: 3035) coordinate system and contains column CELLCODE in the format of 1kmExxxxNxxxx. such as

https://www.eea.europa.eu/data-and-maps/data/eea-reference-grids-2/gis-files/belgium-shapefile

 https://www.eea.europa.eu/data-and-maps/data/eea-reference-grids-2/gis-files/netherlands-shapefile

### (2) Select the grids

select the grids that cover the city and export  the selected features as Grids_NameOfTheCity_Size. gpkg

### (3) Create the mask vector of the city

find a vector of the city and select the features with over 50% covered by the scope of the city through the following process.

add detail

1. find a vector of the city and export the feature as NameOfTheCity.gpkg
2. Vector>>Geoprocessing Tools>>Clip; Input layer: NameOfTheCity.gpkg, Overlay layer: Grids_NameOfTheCity_Size. gpkg from step (2), name the clipped vector as NameOfTheCity_split.gpkg
3. Create a column ‘area’ of NameOfTheCity_split.gpkg by typing in‘=$Aera’
4. Select the features of NameOfTheCity_split.gpkg by “Aera”≥0.5, and export as NameOfTheCity_split_mask.gpkg
5. Vector>>Geoprocessing Tools>>Intersection; 
Input layer: Grids_NameOfTheCity_Size. gpkg
Overlay layer: NameOfTheCity_split_mask.gpkg from step (iv), name the intersection vector as Grids_NameOfTheCity_mask. gpkg(2) Sele
